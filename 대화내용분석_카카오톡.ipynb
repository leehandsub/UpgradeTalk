```
# 날짜로 변환
df.Date = pd.to_datetime(df.Date)

# 파생변수 추가
## 년월일,요일 
df["year"] = df['Date'].dt.strftime('%Y')
df["month"] = df['Date'].dt.strftime('%m')
df["day"] = df['Date'].dt.strftime('%d')
df["weekday"] = df['Date'].dt.day_name()

## 24시간제 표기
df["24time"] = df["timetype"] + " " + df["time"]
df["24time"] = df["24time"].map(lambda x : x.replace("오전","AM"))
df["24time"] = df["24time"].map(lambda x : x.replace("오후","PM"))

temp = []
transform_time = []
for i in range(len(df)) :
    time = df["24time"][i]
    temp.append(dt.datetime.strptime(time,"%p %I:%M"))
    transform_time.append(temp[i].time())

df["24time"] = transform_time

## 글자 수
title_len = []

for i in range(len(df)):
    ttl = len(df['contents'][i])
    title_len.append(ttl)

df['length'] = title_len
df.head()

## 계절 설정
quarter = []
for i in range(len(df)) :
    a = int(df["month"][i])
    if a >= 1 and a <= 3 :
        quarter.append("1q")
    if a >= 4 and a <= 6 :
        quarter.append("2q")
    if a >= 7 and a <= 9 :
        quarter.append("3q")
    if a >= 10 and a <= 12 :
        quarter.append("4q")

df["quarter"] = quarter

# 데이터 확인
df.head()
```
